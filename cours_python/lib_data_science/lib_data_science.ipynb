# NumPy

## Arrays

### Qu'est-ce qu'un array ?

Array = tableau NumPy => liste homogéne, qui ne peut contenir qu'un seul type de donnée.

NumPy dnne accés à des fonctions analytiques.

L'homogénéité des arrays permet de gagner en mémoire et en performance sur les opéraitons.

documentation NumPy : https://docs.scipy.org/doc/numpy-1.13.0/reference/ufuncs.html

```python
import numpy as np
```

```python
revenus = [1800, 1500, 2200, 3000, 2172]
print("La moyenne: ", np.mean(revenus))

x = [-2, -1, 1, 2]
print("La valeur absolue: ", np.abs(x))
print("Exponentielle: ", np.exp(x))
print("Logarithme: ", np.log(np.abs(x)))
```

### Création d'un array

Peu être généré à partir d'une liste Python classique :

```python
revenus_array = np.array(revenus)
revenus_array
```

Si dans la liste de départ il y a des données de types différents, NumPy essaiera de tous les convertir au type le plus général.<br>
Par exemple, si un tableau contient des entiers (  int  ) et des nombres décimaux (float ), tous ses éléments seront convertis en nombres décimaux (float ).

Il existe également des fonctions NumPy permettant de créer des arrays selon un certain pattern ou une spécification particulière. Les plus couramment utilisées sont :
- np.zeros(n): permet de créer un array rempli de 0, de n éléments ;
- np.ones(n) : similaire en tous points à la fonction présentée ci-dessus, à la différence que l’array sera rempli cette fois-ci de 1 ;
- np.arange(i, j, p)  : permet de créer un array rempli avec une séquence linéaire, qui ira de  i  à  j, par pas de  p  ;
- np.linspace(i, j, n) : permet de créer un array de  n  valeurs espacées uniformément entre  i  et  j.

Il est possible d'obtenir le type d'un array via la méthode <b>.dtype</b>.<br>
Les types en NumPy seront toujours présentés via deux informations : le type (i.e: int) et la précision (i.e: 32) qui est le nombre de bits sur lesquels sont codés les nombres.<br>
Plus d'informations sur les types NumPy disponibles sur la page : https://numpy.org/doc/stable/user/basics.types.html

### Sélectionner des éléments d'un array

Comme pour les listes, se fait via l'indice de l'élément

```python
print("5éme élément : ", revenus_array[4])
print("3 premiers éléments : ", revenus_array[0 : 3])
print("3 premiers éléments : ", revenus_array[: 3])
print("Les éléments à partir de l'indice 2: ", revenus_array[2 :])
print("Tous les éléments avec pas de 2 : ", revenus_array[:: 2])
print("La lsite à l'enver : ", revenus_array[:: -1])
```

Il est également possible d'accéder à un élément ou plus selon une condition ou une condition avec la syntaxe <b>mon_array[condition]</b>.<br>
Il est possible de passer plusieurs conditions, à condition que chacune soit entre parenthése et séparé par <b>&</b> (et) ou <b>|</b> (ou).

```python
print("Revenus > 2000", revenus_array[revenus_array > 2000])
print("1000 <= revenu < 2000", revenus_array[(revenus_array >= 1000) & (revenus_array < 2000)])
```

Ces syntaxes (indice ou conditionnelle) ne permettent pas uniquement de sélectionner des éléments, mais également de les modifier.

```python
revenus_array[3] = 1790
revenus_array[revenus_array > 2000] = 0

revenus_array
```

## Tableaux

Array est un objet multidimensionnel.

Il est possible de transformer une liste de listes (2D) classique Python en array multidimensionnel via la méthode <b>np.array(tableau)</b> (Possible de faire plus de 2D).

```python
tab1 = np.array([[1, 2],
                 [3, 4],
                 [5, 6]])

tab2 = np.array([[1, 2, 3],
                 [4, 5, 6]])

print("tableau 1 : ", tab1)
print("tableau 2 : ", tab2)
```

On peut également utiliser des fonctions NumPy pour créer des tableaux plus ou moins complexes.

Exemples de fonctions couramment utilisées :

```python
# un tableau de 3x5 rempli de 1
np.ones((3, 5))

# un tableau de 4 lignes et de 4 colonnes contenant que des 0
np.zeros((4, 4))

# un tableau de 6x3 rempli de valeurs aléatoires comprises entre 0 et 1
np.random.random((6, 3))

# un tableau de 3x3 rempli de valeurs aléatoires entières, comprises entre 1 et 10
np.random.randint(1, 10, size=(3, 3))
```

## Agréger des données avec Pandas

### Agréger plusieurs lignes

Avec la méthode Pandas : <b>.groupby()</b>

Cette méthode permet d'agréger plusieurs lignes par une ou plusieurs colonne (= index du résultat agrégé) suivant une <b>fonction d'agrégation</b>.

L’index va permettre de créer plusieurs groupes : un pour chaque valeur unique de l’index. La fonction va retourner une unique valeur pour chacun des groupes.

<u>Exemple :</u>
- Pour calculer la moyenne de la variable "col" pour chaque valeur de la variable "fix" dans le dataframe suivant

```python
test = pd.read_csv('sources/test.csv')
display(test)
```

```python
test.groupby('fix').mean()
```

Décomposons ce qu’il se passe lors de l’exécution de cette ligne.

Dans un premier temps, notre index sera la variable "fix" ; il va donc y avoir 2 groupes créés, un pour chaque valeur dans notre variablefix .



Sur chacun de ces groupes, on va appliquer la fonction d’agrégation choisie (ici, la moyenne) afin d’avoir le résultat souhaité :



Application de la fonction mean (moyenne) aux variables a et b.

Exemples sur le dataframe "prets" :
- Calculer le chiffre d'affaires mensuel réalisé par chaque agence :

```python
prets.groupby('ville').sum()
```

On observe que sont créés un "group" par ville et que toutes les autres colonnes sont agrégées.

- Calculer le chiffre d'affaires mensuel réalisé par chaque agence ET par type de prêt (il suffit de passer une liste de colonnes en paramètre de .groupby) :

```python
prets.groupby(['ville', 'type']).sum()
```

On observe des sous-groupe "type" à l'intérieur des groupes "ville". L'ordre des variable dans la fonction .gropuby détermine l'organisation en groupe / sous-groupes.

- N'avoir le résultat que sur la variable remboursement :

```python
prets.groupby(['ville', 'type'])['remboursement'].sum()
```

- Calculer la moyenne et la somme deremboursement par agence, ainsi que le maximum derevenu  :

```python
prets.groupby('ville').agg({'remboursement': ['sum', 'mean'], 'revenu': 'max'})
```

Comme montré ci-dessus, il est possible d'appliquer des fonctions d’agrégation différentes en fonction de la colonne, voire appliquer plusieurs fonctions d’agrégation sur une même colonne.
<br><br>

<u>Remarque :</u>
Lorsqu’on utilisegroup_by , les variables fixées en index se retrouvent en index. Si on veut essayez d’accéder à la variable ville du résultat du group by, on aura une erreur ! Il faut effectuer un <b>reset_index</b> pour la replacer en “colonne”.

### Agréger des lignes et des colonnes

L'idée ici est de travailler sous forme d'un tablau à double entrée.

Si on reprends l'exemple précédent, il s'agirait d'avoir dans notre agrégation : en ligne les agences et en colonne les différents types de prêts.

La méthode utilisée est : <b>.pivot_table()</b>, elle prend 4 paramétres :
- <b>index</b> : variable(s) placée(s) en ligne;
- <b>columns</b> : variable(s) placée(s) en colonne(s);
- <b>values</b> : variable sur laquelle on va appliquer la fonction d'agrégation;
- <b>aggfunc</b> : fonction d'agrégation.

```python
data = prets.pivot_table(index='ville', columns='type', values='remboursement', aggfunc='sum')
data
```

```python
# Reset des index pour pouvoir manipuler data comme un dataframe
data = data.reset_index()
data
```

La méthode permettant la transformation inverse s'appelle <b>.melt()</b>. Elle prends des informations sur plusieurs colonnes dans un tableau à double entrée pour le rendre plus "linéaire".

<b>.melt()</b> prends plusieurs paramètres dont :
- <b>id_vars</b>, variable sur laquelle on agrége les données;
- <b>value_vars</b>, liste des colonnes que l'on veut "linéariser".

```python
data.melt(id_vars='ville', value_vars=['automobile', 'immobilier'])
```

## Fusionner des données

Permet de relier plusieurs dataframe et mettre les données en corrélation = jointure.<br>
La jointure se fait via une clé, présente dans chacun des dataframe a joindre,  composer de une ou plusieurs colonnes.

### Fusion avec <b>merge</b>

La jointure se fait avec la fonction <u>ou</u> méthode <b>merge</b>. 2 syntaxes sont possibles :
- Avec la fonction : pd.merge(A, B);
- Avec la méthode : A.merge(B).
Les deux syntaxes sont équivalente : même résultats et arguments a paramétrer.

Par défaut, si la clé n'est pas précisée, Pandas va chercher les colonnes en commun dans les différents data frames (celles qui portent le même nom) et va les sélectionner comme clés = jointure naturelle.  

Dans le cas où l'on spécifie la clé, il va donc y avoir deux cas possibles :
- lorsque la clé porte le même nom dans chaque data frame, on peut utiliser l’argument <b>on</b> ;
- lorsque la clé n’a pas le même nom, il est nécessaire de préciser quelle clé on va utiliser dans chaque data frame. On aura alors besoin des arguments <b>right_on</b> et <b>right_on</b>, qui correspondront à la clé dans le data frame à droite et celle dans le data frame à gauche.


<u>Exemples :</u>
- jointure entre 2 dataframes A et B :<br>
    - <b>pd.merge(A, B, on='id')</b>
- jointure entre 2 dataframes A et C :<br>
    - <b>C.merge(A, left_on='identifiant', right_on='id')</b>
- jointure entre 2 dataframes A et B :<br>
    - <b>pd.merge(A, B, left_on='id', right_on='identifiant')</b>

Décomposons chaque cas :
- Dans le premier cas, c’est une jointure simple sur une clé "id". Cela suppose donc qu’il y a dans le data frame A et dans le data frame B, une colonne qui porte ce nom. Cela renverra une erreur dans le cas contraire ;
- Dans le second cas, le data frame considéré comme "left" est le data frame sur lequel on effectue la jointure : ici, il s’agit donc du data frame C. Le data frame A est considéré comme étant à droite. La jointure est réalisée sur la clé "identifiant" pour le data frame de gauche (ici, C), et id pour l’autre ;
- Dans le dernier cas, le data frame considéré comme "left" est le premier renseigné, donc ici A. B est le data frame "right". La jointure est réalisée avec la clé "id" du data frame de gauche (ici, A), et "identifiant" du data frame de droite.

<br>

<u>Exemple :</u><br>
Jointure entre les dataframes <b>prets</b> et <b>clients</b> dans le nouveau dataframe <b>data</b>,en utilisant la clé commune <b>identifiant</b> :

### Différents types de jointures

Le type de jointure permet de déterminer comment Pandas doit traiter les différentes clé des dataframes, notamment en cas de non correspondance : clé présente dans un dataframe mais pas dans l'autre.

Il existe 4 types :
- Interne (inner join) : type de jointure par défaut, ne conserve que les lignes dont la clé est à droite <b>ET</b> à gauche;
- Gauche (left join) : jointure conservant <b>obligatoirement</b> les clé de gauche et qui compléte avec les éléments de droite disponible. Pour les valeur manquante la valeur par défaut (<b>NaN</b>) est insérée;
- Droite (right join) : équivalent de <b>left join</b> mais avec la table de droite en principale;
- Externe (outer join) : conjonction des jointure <b>left join</b> et <b>right join</b>, converve <b>toutes</b> les clé trouvées à gauche ou à droite en complétant les données manquantes avec des valeurs par défaut.

<br>
<u>Exemple :</u>

Prenons les datafrales "taille et "poids" suivants :<br><br>
 

```python
tailles = pd.read_csv('sources/tailles.csv')
display(tailles)
```

```python
poids = pd.read_csv('sources/poids.csv')
display(poids)
```

```python
# Inner join
pd.merge(tailles, poids, on='id', how='inner')
```

```python
# left join
pd.merge(tailles, poids, on='id', how='left')
```

```python
# right join
pd.merge(tailles, poids, on='id', how='right')
```

```python
# outer join
pd.merge(tailles, poids, on='id', how='outer')
```

### Concaténation

La concaténation permet d'assembler plusieurs dataframe ayant la même structure, les même colonnes et même types.

La fonction Pandas permettant la concaténation est <b>.concat()</b>, qui prend une liste de dataframes en paramètre :
- pd.concat([df1, df2, ..., dfn])

La concaténation ne gére pas les index par défaut. L'opération met juste les dataframes à la suite des autres, sans modification. Il y aura donc des doublons.<br>
Pour éviter les doublons il est possible d'utiliser l'argument <b>ignore_index</b>, équivalent à appliquer la méthode <b>reset_index</b> :
- pd.concat([df1, df2, ..., dfn], ignore_index=True)

### Nuage de points

<u>Exemple :</u>

Représentation du taux d'endettement en fonction du revenu.

Ici les deux variables sont numériques, sans évolution dans le temps, le nuage de point (<b>scatter()</b>) est donc parfaitement adapté pour visualiser la répartition.

<b>scatter()</b> nécessite de définir obligatoirement <b>x</b> et <b>y</b>, qui sont les valeurs à placer en abscisse et en ordonnée :

```python
# Ajout de la variable "taux_endettement" dans le dataframe
prets['taux_endettement'] = round(prets['remboursement'] * 100 / prets['revenu'], 2)

# Création du nuage de points
plt.scatter(prets['revenu'], prets['taux_endettement'])
```

Il est évidement possible de personnaliser un nuage de point en ajoutant des argument, par exemple :
- la couleur des points avec <b>couleur</b>  ou <b>c</b> ;
- la taille des points avec <b>size</b> ou <b>s</b> ;
- le type de marker avec <b>marker</b> ;
- la transparence des points avec <b>alpha</b>.

Exemple avec le même graphique mais personnalisé :

```python
plt.scatter(prets['revenu'], prets['taux_endettement'], s=60, alpha=0.5, c='red', marker='P')
```

D'autres arguments existe.

Plus d'information sur <b>scatter()</b> : https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.scatter.html

### Diagramme circulaire

La fonction pour tracer un diagramme circulaire est <b>pie()</b>.

Son utilisation est similaire à <b>scatter()</b>, il y deux arguments obligatoires à préciser :
- <b>labels</b>, variable non numérique sur laquelle on été agrégées les données ;
- <b>x</b>, les valeurs agrégées correspondantes.

```python
# Agrégation des données
data = prets.groupby('ville')['remboursement'].sum()
data = data.reset_index()

# Création du diagramme circulaire
plt.pie(x=data['remboursement'], labels=data['ville'])
```

Le graphique peut être modifié et amélioré avec des arguments complémentaires.<br>
Par exemple, il est possibl d'afficher le pourcentage associ à chaque "part" avec l'argument <b>autopct</b>.

```python
plt.pie(x=data['remboursement'], labels=data['ville'], autopct='%.2f%%')
```

'%.2f%%' : permet d'afficher la part (pourcentage) pour le label, avec deux décimales et suivit d'un symbole '%'.

Plus d'information sur les diagrammes circulaire dans la documentation officielle : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

### Diagramme à barres

Alternative du diagramme circulaire. On peut représenter exactement la même information, mais avec une perspective différente.

La fonction dédiée est : <b>bar()</b>, 2 arguments sont obligatoires :
- <b>x</b>, valeurs de la variable non numérique (équivalent de l'argument <b>labels</b> de <b>pie()</b>) ;
- <b>height</b>, valeurs agrégées (éqivalent de l'argument <b>x</b> de <b>pie</b>).

En reprennant la même information que pour le diagramme circulaire :

```python
plt.bar(height=data['remboursement'], x=data['ville'])
```

Matplotlib ne permet toutefois pas nativement de réaliser un ordonnencement des données, il faudra donc trier le dataframe en amont :

```python
data_sorted = data.sort_values('remboursement', ascending=True)
plt.bar(height=data_sorted['remboursement'], x=data_sorted['ville'])
```

Lien de la documentation officielle : https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.bar.html

### Créer plusieurs graphiques dans une même fenêtre

Exemple : On souhaite, en un seul graphique, avoir une idée du comportement des agences en termes d’attribution de taux, en fonction du taux.

Représenter le revenu en fonction du taux est assez simple : on peut faire cela via un nuage de points. Et l'information de l'agence pourrait être représenté via la couleur des points.

Cependant, il n’existe pas d’option par défaut avec Matplotlib pour mettre en couleur les points. On va être forcés de créer plusieurs graphiques – un par agence – qu’on va superposer sur une seule et même fenêtre graphique.

Pour cela, à partir du dataframe principale, on va créer un dataframe par agence (ville). Puis définir  un nuage de point pour chacun de ces dataframe. Lorsque que le graphique sera généré, tous les nuages de points y seront tracé.

Afin de rendre lisible le graphique, il faut y ajouter une légende. Pour cela il faut utiliser l'argument <b>label></b> pour chacun des nuage de points, puis utiliser la fonction <b>legend()</b> avant de générer le graphique.

```python
df1 = prets.loc[prets['ville'] == 'TOULOUSE', :]
df2 = prets.loc[prets['ville'] == 'PARIS', :]

plt.scatter(df1['revenu'], df1['taux'], label='TOULOUSE')
plt.scatter(df2['revenu'], df2['taux'], label='PARIS')

plt.legend()
plt.show()
```

Cependant, générer les dataframes peut être fastidieux et source d'erreur. Cela peut être résolu via une boucle.

```python
for ville in prets['ville'].unique():
    df = prets.loc[prets['ville'] == ville, :]
    plt.scatter(df['revenu'], df['taux'], label=ville)
plt.legend()
plt.show()
```

### Personnalisation des graphiques

La personnalisation va permettre de rendre les graphiques conformes aux bonnes pratiques.<br>
En ajoutant des éléments tels que le titre, les lablels des axes, une légende, <i>etc</i>.

#### Modifier les éléments extérieurs

Prenons par exemple le graphique les remboursements (~chiffres d'affaire) totaux par agence :

```python
plt.bar(height=data['remboursement'], x=data['ville'])
```

On observe qu'il manque quelques grilles de lecture :
- On ne sait pas à quoi correspond l’axe des ordonnées, ni quelle est l’unité utilisée ;
- Il manque l’information globale de ce qui est représenté.

Il faut donc ajouter un titre global et au moins le titre de l’un des axes. Pour ce faire, il faut appliquer des fonctions supplémentaires à la fenêtre graphique.

```python
plt.bar(height=data['remboursement'], x=data['ville'])
plt.title("Chiffre d'affaire réalisé par agence", fontname='Arial', fontsize=18)
plt.xlabel("Agences", color='red', fontweight='bold')
plt.ylabel("chiffre d'affaire (€)")
```

#### Modifier les éléments intérieurs

On a ainsi un graphique complet avec une grille de lecture claire. Il existe cependant d’autres options pour donner plus de clarté à une visualisation, ou pour en améliorer la qualité esthétique.

En effet, il y a de nombreux aspects internes au graphique sur lesquels on peut jouer. Nous avons déjà vu l’affichage textuel des valeurs sur les pie plots ou les barplots. Il est également possible de jouer – entre autres – sur les graduations (ou ticks, en anglais), le quadrillage, ou encore les couleurs de fond.

Voici un exemple :

```python
plt.figure(figsize=(10,6))
plt.bar(height=data['remboursement'], x=data['ville'])
plt.title("Chiffre d'affaire réalisé par agence", fontname='Arial', fontsize=18)
plt.xlabel("Agences", color='red', fontweight='bold')
plt.ylabel("chiffre d'affaire (€)")
plt.grid(axis='y')
plt.ylim(0, 120000)
plt.yticks([0, 30000, 60000, 90000, 120000])

ca = data['remboursement'].to_list()
for i in range(6):
    plt.text(i -0.2, ca[i] + 2000, round(ca[i], 1))

plt.show()
```

### Présentation

Seaborn = librairie surcouche à Matplotlib.

Seaborn corrige plusieurs défaut de Matplotlib :
- Il propose de multiples modèles graphiques prédéfinis de bonne qualité esthétique, en modifiant les options graphiques par défaut de Matplotlib;
- Il ajoute une interaction avec les data frames afin de faciliter grandement la génération de graphiques à partir de ceux-ci;
- Il propose un catalogue – très – dense de fonctions graphiques pour répondre le plus précisément possible à une problématique donnée.

En contrepartie, Seaborn propose un peu moins d’options de personnalisation en tant que telles.<br>
Mais Seaborn étant une surcouche de Matplotlib, il est possible d’utiliser l’ensemble des options de personnalisation de Matplotlib, directement sur des graphiques tracés avec Seaborn.

<u>Exemple :</u>

Nuage de point avec Seaborn : <b>scatterplot</b>, qui a prend obligatoirement en paramètres :
- <b>data</b> = dataframe utilisé pour tracer le graphique ;
- <b>x</b> et <b>y</b> = variables du dataframe utilisées respectivement en absisce et en ordonnée.

```python
sns.scatterplot(data=prets, x='revenu', y='taux_endettement')
```

Le réel intérêt de Seaborn apparaît lorsqu’on veut d’ajouter des dimensions supplémentaires.

Par exemple en reprennant le graphique précédent pour y ajouter en couleur le type de prets. Là où on avait besoin d’une boucle avec Matplotlib, voilà le code avec Seaborn :

```python
sns.scatterplot(data=prets, x='revenu', y='taux_endettement', hue='type')
```

En précisant la variable via l’argument <b>hue</b>, Seaborn crée automatiquement une couleur pour chaque valeur existante.

Si l'on veut, plutôt que l’information du type de prêt, l’information de la durée du prêt. On peut faire cela en associant la taille d’un point à la durée d’un prêt :

```python
sns.scatterplot(data=prets, x='revenu', y='taux_endettement', size='duree')
```

Bien sûr, le nombre d’informations à afficher n’est pas limité, on pourrait très bien mixer la couleur avec la taille des points.

Et comme dit précédement, il est possible d'appliquer des fonctions Matplotlib :

```python
plt.figure(figsize=(10,6))

plt.rcParams.update({'font.size': 14})

sns.scatterplot(data=prets, x='revenu', y='taux_endettement', hue='type', size='duree')

plt.ylabel("Taux d'endettement")

plt.xlabel('Revenu mensuel (€)')

plt.grid()

plt.xlim(500, 7500)

plt.legend(bbox_to_anchor=(1, 1.02))

plt.title("Taux d'endettement en fonction du revenu mensuel\npar type de prêt contracté")

plt.show()
```

Explications des différentes fonctions utilisées :
- La fonction <b>rcParams.update</b>, avec l’argument <b>font.size</b>, définit pour tous les graphiques – et non pas uniquement le graphique actuel – la taille de la police de tous les éléments (xlabel, title, legend, etc.) à 14;
- <b>grid</b> permet d’afficher le quadrillage;
- La fonction <b>xlim</b> , fixe les limites des graduations de 500 à 7 500;
- La fonction <b>legend</b> affiche la légende sur le graphique. L’utilisation de l’argument <b>bbox_to_anchor</b> permet de fixer la position de la légende en dehors du graphique, aux coordonnées renseignées.

Autre fonctionnalité importante : l'agrégation.

Avec Matplotlib, avant de tracer, il faut au préalable d’agréger nos données.<br>
La fonction <b>barplot</b> de Seaborn permet de réaliser cette étape directement à l’intérieur du graphique.

<u>Exemple :</u>

Diagramme en barre du chiffred'affaire par agence.

```python
sns.barplot(data=prets, x='ville', y='remboursement', ci=None, estimator=sum)
```

Le début est similaire à la fonction <b>scatterplot</b>, mais présente ensuite deux nouveaux arguments :
- <b>ci</b>, pour confidence interval (ou intervalle de confiance), qui ajoute les intervalles de confiance à chaque barre. Fixer cet argument àNone permet de les faire disparaître;
- <b>estimator</b> correspond à la fonction d’agrégation.

Tous les arguments vus précédemment fonctionnent également, notamment avoir des sous-groupes – par exemple : par type de prêt – en couleur :

```python
sns.barplot(data=prets, x='ville', y='remboursement', ci=None, estimator=np.mean, hue='type')
```

Seaborn possède de nombreuses fonctions (plus d’une vingtaine) pour répondre à toutes les situations possibles en termes de visualisation.

<u>Par exemple :</u>
- <b>lineplot</b> pour tracer des courbes;
- <b>boxplot</b> pour des boîtes à moustaches;
- <b>histplot</b> pour des histogrammes;
- <b>kdeplot</b> pour un graphique à densité;
- <b>pairplot</b> pour tracer l’ensemble des variables numériques deux à deux, sur un seul graphique.
